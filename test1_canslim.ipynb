```python
import pandas as pd
from vnstock3 import Vnstock
```

```python
tickers = ['VNM', 'TCB', 'VIC', 'MSN', 'HPG', 'VPB', 'VHM', 'VJC',
           'MWG', 'MBB', 'SAB', 'VCB', 'HDB', 'FPT', 'STB', 'NVL',
           'EIB', 'VRE', 'PNJ', 'GAS', 'SSI', 'CTG', 'GMD', 'CTD',
           'ROS', 'REE', 'SBT', 'CII', 'DPM', 'DHG']
```

```python
def fetch_financial_data(ticker, period):
    stock = Vnstock().stock(symbol=ticker, source='VCI')
    income_statement = stock.finance.income_statement(period=period, lang='vi')
    finance_ratios = stock.finance.ratio(period=period, lang='vi')
    
    # Extract and merge data
    ratios = finance_ratios['Chỉ tiêu định giá'].copy()  # Create a copy to avoid setting on a slice
    ratios['Năm'] = finance_ratios['Meta']['Năm']
    ratios['Kỳ'] = finance_ratios['Meta']['Kỳ']
    ratios['ROE'] = finance_ratios['Chỉ tiêu khả năng sinh lợi']['ROE (%)']
    
    return pd.DataFrame(income_statement), pd.DataFrame(ratios)
```

```python
# Create dictionaries to store the financial data
data = {ticker: fetch_financial_data(ticker, 'quarter') for ticker in tickers}
yearly_data = {ticker: fetch_financial_data(ticker, 'year') for ticker in tickers}
```

```python
def fetch_average_monthly_volume(ticker, start_date, end_date):
    stock = Vnstock().stock(symbol=ticker, source='VCI')
    df = stock.quote.history(start=start_date, end=end_date, interval='1D')
    df = pd.DataFrame(df)
    df['time'] = pd.to_datetime(df['time'])
    df.set_index('time', inplace=True)
    average_volume_per_month = df['volume'].resample('M').mean().mean()
    return average_volume_per_month
```

```python
average_volume_per_month= fetch_average_monthly_volume('FPT','2017-01-01', '2018-12-31' )
```

```python
def filter_years(df, years):
    return df[df['Năm'].isin(years)].reset_index(drop=True)
```

```python

def calculate_percentage_change(df, column, period):
    df[f'prev_{period}'] = df[column].shift(period)
    df[f'percentage_change_{period}'] = ((df[column] - df[f'prev_{period}']) / df[f'prev_{period}']) * 100
    return df
```

```python
def screen_stocks(tickers, data, yearly_data, years, volume_threshold=300000):
    screened_stocks = []
    
    for ticker in tickers:
        ticker_comp = Vnstock().stock(symbol= ticker, source='TCBS').company.officers()
        ticker_comp_percent = ticker_comp['officer_own_percent'][:5]
        income_state, ratio_state = data[ticker]
        income_state_yearly, ratio_state_yearly = yearly_data[ticker]

        income_state_filtered = filter_years(income_state, years)
        ratio_quarter_filtered = filter_years(ratio_state, years)
        ratio_year_filtered = filter_years(ratio_state_yearly, years)
        
        # Print DataFrames for debugging
        average_volume_per_month  = fetch_average_monthly_volume(ticker, '2017-01-01', '2018-12-31')
        
        
        # Calculate percentage change
        eps_quarter = calculate_percentage_change(ratio_quarter_filtered, 'EPS (VND)', 4)
        revenue_quarter = calculate_percentage_change(income_state_filtered, 'Doanh thu (Tỷ đồng)', 4)
        eps_year = calculate_percentage_change(ratio_year_filtered, 'EPS (VND)', 1)
        
        # Print DataFrames with percentage changes
        # print("EPS Quarter with Percentage Change:")
        # print(eps_quarter.head())
        # print("Revenue Quarter with Percentage Change:")
        # print(revenue_quarter.head())
        # print("EPS Year with Percentage Change:")
        # print(eps_year.head())

        # Check criteria
        if ((eps_quarter['percentage_change_4'].mean() >= 20) and
            (revenue_quarter['percentage_change_4'].mean() >= 25) and
            (eps_year['percentage_change_1'].mean() >= 20) and
            (ratio_state_yearly['ROE'].mean() >= 0.17)) and average_volume_per_month > volume_threshold and (ticker_comp_percent.sum())>=0.1:
            screened_stocks.append(ticker)

    return screened_stocks

# Define the years you are interested in
years_of_interest = [2017, 2018]

# Get screened stocks
screened_stocks = screen_stocks(tickers, data, yearly_data, years_of_interest)
print("Screened stocks:", screened_stocks)
```
